```python
# Library imports
import cadquery as cq
```

```python
# Set variables
height = 2
width = 2.5
thickness = 2
rounded = True
```

```python
# Create lcd screen
lcd_screen = cq.Workplane().rect(width, height).extrude(thickness)
lcd_screen
```

```python
# Add rounded edges if requested
if rounded:
    lcd_screen = lcd_screen.edges("|Z").fillet(0.1)
lcd_screen
```

```python
# export step
cq.exporters.export(lcd_screen, "test_steps/lcd.step")
```
