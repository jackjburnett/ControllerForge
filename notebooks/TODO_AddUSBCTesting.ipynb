```python
# Library imports
import cadquery as cq
```

```python
# Set variables
height = 4
width = 11
corner_radius = 1
wall_thickness = 2
```

```python
# Create the rectangular shape for the USB_C receptacle
usb = cq.Workplane().rect(width, height)
usb = usb.extrude(wall_thickness)
usb
```

```python
# Curve the edges using the corner radius
usb.edges("|Z").fillet(corner_radius)
```

```python
# export step
cq.exporters.export(usb, "test_steps/usb.step")
```
