```python
# Library imports
import cadquery as cq
```

```python
# Locate font
fontPath = 'fonts/LexendDeca.ttf'
```

```python
# Set variables
content = "ABXY"
size = 12
depth = 2
flip = True
```

```python
# Create the text using cadquery
text = cq.Workplane().text(content, size, distance=-depth, font="Arial", halign="center", valign="center")
# text= cq.Workplane().text(content, size, distance=-depth, fontPath=fontPath, halign="center", valign="center")
text
```

```python
# Add the plane to a button for visualisation
test_button = cq.Assembly().add(cq.Workplane().circle(15).extrude(4))
test_button.add(text, loc=cq.Location((0, 0, 4)))
```

```python
# export step
cq.exporters.export(text, "test_steps/text.step")
```
