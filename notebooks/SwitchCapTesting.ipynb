```python
# Library imports
import cadquery as cq
```

```python
# Set variables
button_width = 30.0
button_height = 10.0
top_thickness = 2.0
wall_thickness = 1.0
wall_height= 3.0
bevel=True
# https://cdn.shopify.com/s/files/1/0657/6075/5954/files/SPEC-CPG151101S21_Kailh_Red_Switch_b6770d67-8f51-48d9-9ec1-45442f4de278.pdf?v=1667196539
mount_height= 4.0
mount_radius=3.0
mount_cross_width=4.2
mount_cross_length=1.4
```

```python
# Create the top
top = cq.Workplane().circle(button_width/2)
top = top.extrude(top_thickness)
top
```

```python
# Add bevel
if bevel:
    top=top.edges().fillet(0.99)
top
```

```python
# Create the mount
mount=cq.Workplane().circle(mount_radius)
mount_cross=cq.Workplane().rect(mount_cross_width, mount_cross_length).rect(mount_cross_length, mount_cross_width).extrude(mount_height)
mount=mount.extrude(mount_height)
mount=mount.cut(mount_cross)
mount
```

```python
# Create walls
if wall_thickness>0:
    walls= cq.Workplane().circle(button_width/2).circle((button_width/2)-wall_thickness)
    walls=walls.extrude(wall_height)
    walls
```

```python
# Combine all parts
cap=cq.Assembly().add(top).add(mount, loc=cq.Location((0,0,top_thickness)))
if wall_thickness>0:
    cap.add(walls, loc=cq.Location((0,0,top_thickness/2)))
cap
```

```python
# export stl
cap.save("test_stls/cap.step")
```
