```python
# Define parameters
radius = 10
flattening_factor = 0.5  # This will flatten the sphere to half its height

# Create the top profile
top_profile = cq.Workplane("XY").ellipse(radius, radius * flattening_factor)

# Loft from the bottom to the top profile
oblate_spheroid = cq.Workplane("XY").ellipse(radius, radius * flattening_factor).workplane(offset=radius).ellipse(radius, radius * flattening_factor).loft()

# Render the result
oblate_spheroid
```
